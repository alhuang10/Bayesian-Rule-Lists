```python
# Titanic Data Processing
def convert_class(passenger_class):
    if passenger_class == 1:
        return "First Class"
    elif passenger_class == 2:
        return "Second Class"
    else: 
        return "Third Class"
    
def convert_age(age):
    adult_age_cutoff = 18
    if age >= adult_age_cutoff:
        return "Adult"
    else:
        return "Child"
```

```python
df_no_nan = df.dropna()

df_no_nan['Pclass'] = df_no_nan['Pclass'].apply(convert_class)
df_no_nan['Age'] = df_no_nan['Age'].apply(convert_age)
df_no_nan['Sex'] = df_no_nan['Sex'].apply(lambda x: str.title(x)) # Formatting gender
```

```python
# FP-Tree class

class FP_Node:
    
    # attribute
    # count
    # child list
    
    def __init__(self, attribute, count):
        
        self.attribute = attribute
        self.count = count
        self.child_list = []
        
class FP_Tree:
    
    # item_table - attribute to (count, pointer_to_tree)
    # root - null node
    
    def __init__(self):
        self.root = FP_Node(None, 0)
        self.item_counts = defaultdict(int)
        self.item_nodes = defaultdict(list) # Ordered by insertion into tree

    def insert_list(self, item_list, parent_node):
                
        # Base case
        if len(item_list) == 0:
            return
        else:
            
            attr_to_insert = item_list[0]
            
            # If parent_node has a direct child with same attribute name then increment
            attr_node = next( (node for node in parent_node.child_list if node.attribute == attr_to_insert), None)
            
            # Else create a new node with parent node linked to T
            if attr_node is None:
                
                attr_node = FP_Node(attr_to_insert, 1)
                parent_node.child_list.append(attr_node)
                self.item_nodes[attr_to_insert].append(attr_node) # Add new node to item_nodes
                
            else:
                attr_node.count += 1
            
#             print(self.item_nodes)
            self.item_counts[attr_to_insert] += 1 # Increment item count            
                
            # Recursively call the rest of the list
            # Parent node either a newly created node or an existing child
            self.insert_list(item_list[1:], attr_node)
```
